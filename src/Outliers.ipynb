```python
def log_likelihood(central_value, sigma, obs_data, sigma_obs_data):
    obs_data = sort(obs_data)
    n_points = size(obs_data)
    fake_data = random.normal(loc=central_value,scale=sigma, size=n_points)
    chi_squared = sum(((fake_data - obs_data)**2)/(2.0*sigma_obs_data))
    return -chi_squared
```

```python

def MCMC_walk(n_steps, obs_data, sigma_obs_data, init_center = 0.40, init_sigma=1E-4):
    sigma_walk = empty((0))
    center_walk = empty((0))
    log_l_walk = empty((0))

    

  
    sigma_walk = append(sigma_walk, init_sigma)
    center_walk = append(center_walk, init_center)
    log_l_walk = append(log_l_walk, log_likelihood(init_center, init_sigma, obs_data, sigma_obs_data))

    print log_l_walk

    for i in range(n_steps):
        new_center = random.normal(loc=center_walk[i], scale=5.0E-3)
        new_sigma = random.normal(loc=sigma_walk[i], scale=1.0/c_light)
        while(new_sigma<0):
            new_sigma = random.normal(loc=sigma_walk[i], scale=1.0/c_light)

        log_like_old = log_l_walk[i]
        log_like_new = log_likelihood(new_center, new_sigma, obs_data, sigma_obs_data)
    
        alpha = exp((log_like_new-log_like_old))
        if(alpha>1.0):
            sigma_walk = append(sigma_walk, new_sigma)
            center_walk = append(center_walk, new_center)
            log_l_walk = append(log_l_walk, log_like_new)
        else:
            beta = random.random()
            if(beta<alpha):
                sigma_walk = append(sigma_walk, new_sigma)
                center_walk = append(center_walk, new_center)
                log_l_walk = append(log_l_walk, log_like_new)
            else:
                sigma_walk = append(sigma_walk, sigma_walk[i])
                center_walk = append(center_walk, center_walk[i])
                log_l_walk = append(log_l_walk, log_l_walk[i])
    
    return center_walk, sigma_walk, log_l_walk
```

```python
c_light = 3.0E5
init_center = 0.40
init_sigma = 400.0/c_light
```

```python
c_walk_A, s_walk_A, l_walk_A = MCMC_walk(150000, z_data, sigma_z_data, init_center=init_center, init_sigma=init_sigma)
```
